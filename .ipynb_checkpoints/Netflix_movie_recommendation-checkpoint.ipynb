## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

## Importing dataset

```python
dataset = pd.read_csv('Netflix Movie Recommendation Ec.csv', header = None)
transactions = []
for i in range(0, 7466):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
transactions
```

## Eclat Training on Dataset

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.0003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualizing

### Raw Results

```python
results = list(rules)
```

### Proper Display

```python
def inspect(results):
    movie_1 = [tuple(result[0])[0] for result in results]
    movie_2 = [tuple(result[0])[1] for result in results]
    support = [result[1] for result in results]
    return list(zip(movie_1,movie_2,support))
DataFrame_intelligence = pd.DataFrame(inspect(results),columns = ['movie_1','movie_2','support'])
```

```python
DataFrame_intelligence
```

```python
DataFrame_intelligence.nlargest(n = 15, columns = 'support')
```
